### Imports

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
```

### Read CSV

```python
# salaries.csv
df = pd.read_csv("salaries.csv")
```

```python
df.sample(3)
```

### Split Data

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.drop(columns=["salary_in_usd", "salary"]),
    df['salary_in_usd'],
    test_size=0.2,
    random_state=42
)
```

```python
X_train.head(2)
```

```python
st1 = ColumnTransformer([
    ('work_year', SimpleImputer(strategy='most_frequent'), [0]),
    ('ohe', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), [1,2,3,4,5,7,8])
], remainder='passthrough')
```

```python
st2 = DecisionTreeClassifier()
```

### Pipeline

```python
pipe = Pipeline([
    ('st1', st1),
    ('st2', st2)
])
```

### Train & Predict

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

### Accuracy Score

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

### Export Model

```python
import pickle
```
